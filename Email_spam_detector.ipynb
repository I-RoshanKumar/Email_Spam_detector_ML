```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('spam.csv')
df.head()
```

```python
df.groupby('Category').describe()
```

```python
df['spam']=df['Category'].apply(lambda x:0 if x=='spam' else 1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.Message,df.spam,test_size=0.3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
X_train_count=v.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails=["Hey ram can we get together at the party tonight?",
        'Upon purchase of 3400 you get 45% off.']
emails_count=v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count=v.transform(X_test)
model.score(X_test_count,y_test)
```

# Differnet method by using Pipeline

```python
from sklearn.pipeline import Pipeline
clf=Pipeline([('vectorizer',CountVectorizer()),('nb',MultinomialNB())])
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```
